```python
# 6) What are the top 5 states with the most failed banks?

df.columns  = df.columns.str.strip()
state_failures = df["ST"].value_counts()
top5 = state_failures.head(5)

print("Top 5 states with most failed banks : ",top5)
```

```python
#7) What are the top 5 acquiring institutions?

df.columns = df.columns.str.strip()
acquiring_inst = df["Acquiring Institution"].value_counts()
top5_acqiures = acquiring_inst.head(5)

print("Top 5 acquiring institutions are : ",top5_acqiures)
```

```python
# 8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?

# rows where "acquring instituion is state bank of  texas"
sbt_acquisitions = df[df["Acquiring Institution"] == 'State Bank of Texas']

#how many banks were acquired
total_acquired = len(sbt_acquisitions)

#out of those ,how many were in texas(st == "tx")
acquired_in_tx = sbt_acquisitions[sbt_acquisitions["ST"] == "TX"]
total_in_texas = len(acquired_in_tx)

print("total banks acquired by state bank of texas:",total_acquired)
print("Number of those located in texas:",total_in_texas)
```
